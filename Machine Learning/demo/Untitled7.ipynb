```python
h2o.init()
```

```python
train_data = pd.read_csv("train.csv", sep = ';')
test_data = pd.read_csv("test.csv", sep = ';')
```

```python
def stratify(df, seed=42):
    stratify = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    df['fold'] = np.zeros(len(df), dtype='int')
    
    for (i, (train_idx, val_idx)) in enumerate(stratify.split(df, y=df['quality'])):
        df['fold'].iloc[val_idx] = i
        
    return df
```

```python
train = h2o.H2OFrame(train_data)
test = h2o.H2OFrame(test_data)
```

```python
train_data_red = train_data[train_data['type'] == 'red']
train_data_white = train_data[train_data['type'] == 'white']
```

```python
aml = H2OAutoML(max_models=20, max_runtime_secs=3600) 
```

```python
aml.train(x=X, y=Y, training_frame=train)
```
