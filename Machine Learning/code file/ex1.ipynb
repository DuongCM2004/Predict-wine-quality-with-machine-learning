```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import boxcox
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (StandardScaler, MinMaxScaler,
                                   PolynomialFeatures)
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv', sep = ';')
df.head()
```

```python
test = pd.read_csv('test.csv', sep = ';')
test
```

```python
df1 = df.copy()
test1 = test.copy()
```

```python
df1.drop(columns = ['type'], inplace = True)
df1
```

```python
test1.drop(columns = ['type'], inplace = True)
test1
```

```python
df1.describe()
```

```python
test1.describe()
```

```python
test1['id'] = test1['id'].astype(float)
```

```python
id = test1['id']
test2 = test1.copy()
test2.drop(columns = 'id', inplace = True)
test2
```

```python
data_col = 'quality'
X_train = df1.drop(data_col, axis = 1)
Y_train = df1[data_col]
```

```python
mm = MinMaxScaler()
X_train_s = mm.fit_transform(X_train)
X_test_s = mm.transform(test2)
```

```python
lr = LinearRegression()
lr = lr.fit(X_train_s, Y_train)
```

```python
predicted = lr.predict(X_test_s)
```

```python
id = test1['id'].astype(int)
```

```python
data = {'id': id, 'quality': predicted}
```

```python
data = pd.DataFrame(data)
data
```

```python
csv_filename = 'results.csv'
data.to_csv(csv_filename, index=False)
print(f'DataFrame saved to {csv_filename}')
```
