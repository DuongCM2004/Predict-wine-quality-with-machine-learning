```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression, Lasso, Ridge, SGDRegressor
from scipy.stats import boxcox
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import (StandardScaler, MinMaxScaler,
                                   PolynomialFeatures)
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, cross_val_predict, train_test_split, GridSearchCV
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv', sep =';')
df
```

```python
df1 = df.drop_duplicates()
df1
```

```python
test = pd.read_csv('test.csv', sep =';')
test.head()
```

```python
id = test['id']
test = test.drop(columns = ['id'])
test
```

```python
X_test = test.drop(columns = 'type')
```

```python

one_hot_encoded_data = pd.get_dummies(df1, columns = ['type'])
one_hot_encoded_data
```

```python
y = one_hot_encoded_data['quality']
X = one_hot_encoded_data.drop(columns = ['quality','type_red','type_white'])
```

```python
for column in X.columns:
    X[column] = boxcox(X[column])[0]
```

```python
poly = PolynomialFeatures(degree = 2, include_bias = False)
ss = StandardScaler()
X_train_s = ss.fit_transform(X)
X_test_s = ss.transform(X_test)
X_train_poly = poly.fit_transform(X_train_s)
X_test_poly = poly.transform(X_test_s)
```

```python
lr = LinearRegression()
lr.fit(X_train_poly, y)
```

```python
predicted2 = lr.predict(X_test_poly)
```
